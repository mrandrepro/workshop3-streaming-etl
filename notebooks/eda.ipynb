Limpieza y Harmonización

CARGAR ARCHIVOS CON AÑOS

```python
import pandas as pd
import numpy as np

files = {}
for year in range(2015, 2020):
    df = pd.read_csv(f"../data/raw/{year}.csv")
    df["year"] = year
    files[year] = df
```

FUNCION DE HARMONIZACION


```python
def harmonize(df, year):
    if year in [2015, 2016]:
        df = df.rename(columns={
            "Country":                        "country",
            "Happiness Score":                "happiness_score",
            "Economy (GDP per Capita)":       "gdp",
            "Family":                         "family",
            "Health (Life Expectancy)":       "health",
            "Freedom":                        "freedom",
            "Generosity":                     "generosity",
            "Trust (Government Corruption)":  "corruption",
        })
    elif year == 2017:
        df = df.rename(columns={
            "Country":                        "country",
            "Happiness.Score":                "happiness_score",
            "Economy..GDP.per.Capita.":       "gdp",
            "Family":                         "family",
            "Health..Life.Expectancy.":       "health",
            "Freedom":                        "freedom",
            "Generosity":                     "generosity",
            "Trust..Government.Corruption.":  "corruption",
        })
    elif year in [2018, 2019]:
        df = df.rename(columns={
            "Country or region":              "country",
            "Score":                          "happiness_score",
            "GDP per capita":                 "gdp",
            "Social support":                 "family",
            "Healthy life expectancy":        "health",
            "Freedom to make life choices":   "freedom",
            "Generosity":                     "generosity",
            "Perceptions of corruption":      "corruption",
        })

    cols = ["country", "year", "happiness_score",
            "gdp", "family", "health",
            "freedom", "generosity", "corruption"]
    return df[cols]
```

APLICAR Y UNIR

```python
dfs = []
for year, df in files.items():
    df_clean = harmonize(df, year)
    dfs.append(df_clean)

unified = pd.concat(dfs, ignore_index=True)
print(f"Shape: {unified.shape}")
unified.head(10)
```

VERIFICAR VALORES NULOS  Y DUJPLICADOS

```python
print("Nulos por columna:")
print(unified.isnull().sum())

print(f"\nDuplicados: {unified.duplicated().sum()}")
```

LIMPIAR NULOS Y DUPLICADOS

```python
# Forzar tipos numéricos (corruption en 2018 a veces viene como string)
numeric_cols = ["happiness_score", "gdp", "family",
                "health", "freedom", "generosity", "corruption"]

for col in numeric_cols:
    unified[col] = pd.to_numeric(unified[col], errors="coerce")

# Eliminar filas con nulos
before = len(unified)
unified = unified.dropna()
after = len(unified)
print(f"Filas eliminadas por nulos: {before - after}")

# Eliminar duplicados
unified = unified.drop_duplicates()
print(f"Shape final: {unified.shape}")
```

se eliminan las filas con nulos porque representan menos del 1% del dataset y su imputación podría introducir sesgo en el modelo. Se eliminan duplicados para evitar que el modelo sobreponderen ciertos países.X

VALIDAR Y GUARDAR

```python
print(unified.dtypes)
print("\nEstadísticas finales:")
print(unified.describe())

unified.to_csv("../data/processed/unified.csv", index=False)
print("\nunified.csv guardado en data/processed/")
```
